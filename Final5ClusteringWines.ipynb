## Piedmont wines dataset
- There are 178 samples of distinct variety of red wine from the Piedmont region of Italy.
- The features measures chemical composition, e.g. alcohol content
- .. and also visual properties like "Colorintensity"

## Load the dataset

```python
import pandas as pd
import numpy as np
```

```python
col = ["Alcohol",
"Malic_acid",
"Ash",
"Alcalinity_of_ash",
"Magnesium",
"Total_phenols",
"Flavanoids",
"Nonflavanoid_phenols",
"Proanthocyanins",
"Colorintensity",
"Hue",
"OD280/OD315_of_diluted_wines",
"Proline"]
df = pd.read_csv('datasets/wine.data.txt', names=col)
df.head()
```

```python
# The varieties of the wines (there are 3 different types)
ground_truth = ['Barolo']*59+['Grignolio']*71+['Barbera']*48
ground_truth
```

# Clustering the wines
The code below creates a kmeans model with n_clusters =3. But the crosstab shows that the clustering result is awful:

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3)
labels = model.fit_predict(df)
labels
```

```python
df_ct = pd.DataFrame({'predicted':labels, 'ground_truth':ground_truth})
ct = pd.crosstab(df_ct['predicted'], df_ct['ground_truth'])
ct
```

The crosstab suggests that our clustering model did not work well. What can we do to improve the model? Show your work below and your crosstab after applying your fix:

(hint: with one fix, we can improve the performance of the model a lot, making the result almost perfect)

<font color='red'>Not much will change based on the model parameters, it has to do with the data itself. Looking at the data, the values vary for all features. Scaling the data in the data set should allow for better and more accurate clustering. </font>

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df)
df =scaler.transform(df)
```

```python
df1 = pd.DataFrame(df)
df1.head()
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=3,random_state=42,n_init=50,max_iter=900).fit(df)
labels = model.fit_predict(df)
labels
```

```python
df_ct = pd.DataFrame({'predicted':labels, 'ground_truth':ground_truth})
ct = pd.crosstab(df_ct['predicted'], df_ct['ground_truth'],margins=True, margins_name='Total')
ct
```
